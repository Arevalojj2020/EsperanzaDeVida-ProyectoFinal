```python
import wbgapi as wb
import pandas as pd
import numpy as np
import time

def getDataFrameDatabank(paises_lista, diccionario_de_datos):
    dataset = np.array([])

    for i in diccionario_de_datos.items():
        test = wb.data.fetch(i[0], paises_lista)
        time.sleep(1)
        lista_de_paises = []
        lista_de_años = []
        lista_de_indicadores = []

        for a in test:
            year = int(a['time'][2:])
            if year >= 2022 and year <= 2022:
                lista_de_paises.append(a['economy'])
                lista_de_años.append(year)
                lista_de_indicadores.append(a['value'])

        add = {'title': i[1], 'economy': lista_de_paises, 'time': lista_de_años, 'value': lista_de_indicadores}
        dataset = np.append(dataset, add)

    for i in range(0, len(dataset)):
        print('dimension of data list: ' + str(dataset[i]['title']))
        print(len(dataset[i]['value']))

    for i in range(0, len(dataset) - 1):
        if dataset[i]['time'] == dataset[i + 1]['time']:
            print('is equal times')
        if dataset[i]['economy'] == dataset[i + 1]['economy']:
            print('is equal countries')

    tabla_indicadores = pd.DataFrame()
    tabla_indicadores['country'] = dataset[0]['economy']
    tabla_indicadores['year'] = dataset[0]['time']

    for i in range(0, len(dataset)):
        years = dataset[i]['time']
        values = dataset[i]['value']
        filtered_years = []
        filtered_values = []

        for j in range(len(years)):
            year = int(years[j])
            if year >= 2022 and year <= 2022:
                filtered_years.append(years[j])
                filtered_values.append(values[j])

        if len(filtered_values) == 0:
            tabla_indicadores[dataset[i]['title']] = np.nan
        else:
            tabla_indicadores[dataset[i]['title']] = filtered_values

    return tabla_indicadores

paises_lista = ['DEU', 'AUS', 'CAN', 'ESP', 'USA', 'ITA', 'JPN', 'GBR', 'RUS', 'SWE',
                'ALB', 'ARG', 'BRA', 'COL', 'VNM', 'PHL', 'IND', 'IDN', 'MEX', 'PER',
                'ZAF', 'THA', 'AFG', 'BGD', 'EGY', 'ETH', 'IRQ', 'IRN', 'KEN', 'MAR']

diccionario_de_datos = {'SP.DYN.LE00.IN': 'Life expectancy at birth, total (years)',
    'NY.GDP.PCAP.CD': 'GDP per capita (current US$)',
    'NY.GNP.MKTP.CD': 'GNI (current US$)',
    'NY.GDP.MKTP.CD': 'GDP (current US$)',
    'SP.URB.TOTL.IN.ZS': 'Urban population (% of total population)',
    'SP.RUR.TOTL.ZS': 'Rural population (% of total population)',
    'MS.MIL.XPND.GD.ZS': 'Military expenditure (% of GDP)',
    'NE.CON.GOVT.CD':'General government final consumption expenditure (current US$)',
    'SP.POP.GROW':'Population growth (annual %)',
    'EG.ELC.ACCS.ZS': 'Access to electricity (% of population)',
    'TX.VAL.FOOD.ZS.UN':'Food exports (% of merchandise exports)',
    'AG.PRD.FOOD.XD':'Food production index (2014-2016 = 100)',
    'NE.CON.PRVT.ZS': 'Households and NPISHs final consumption expenditure (% of GDP)'}

tabla_indicadores2 = getDataFrameDatabank(paises_lista, diccionario_de_datos)
```

```python
# Valor para rellenar la columna

Life_expectancy_at_birth_total_years = 71.71
GDP_per_capita_current_US = 13248.38
GNI_current_US = 1217376871771.15
GDP_current_US = 1204602588401.44
Urban_population_of_total_population = 60.39
Rural_population_of_total_population = 39.61
Military_expenditure_of_GDP = 1.98
General_government_final_consumption_expenditure_current_US = 209676319230.43
Population_growth_annual = 1.30
Access_to_electricity_of_population = 87.97
Food_exports_of_merchandise_exports = 16.57
Food_production_index_2014_2016_100 = 84.77
Households_and_NPISHs_final_consumption_expenditure_of_GDP = 62.11

# Rellenar las columnas con un solo valor
tabla_indicadores2['Life expectancy at birth, total (years)'] = tabla_indicadores2['Life expectancy at birth, total (years)'].fillna(Life_expectancy_at_birth_total_years)
tabla_indicadores2['GDP per capita (current US$)'] = tabla_indicadores2['GDP per capita (current US$)'].fillna(GDP_per_capita_current_US)
tabla_indicadores2['GNI (current US$)'] = tabla_indicadores2['GNI (current US$)'].fillna(GNI_current_US)
tabla_indicadores2['GDP (current US$)'] = tabla_indicadores2['GDP (current US$)'].fillna(GDP_current_US)
tabla_indicadores2['Urban population (% of total population)'] = tabla_indicadores2['Urban population (% of total population)'].fillna(Urban_population_of_total_population)
tabla_indicadores2['Rural population (% of total population)'] = tabla_indicadores2['Rural population (% of total population)'].fillna(Rural_population_of_total_population)
tabla_indicadores2['Military expenditure (% of GDP)'] = tabla_indicadores2['Military expenditure (% of GDP)'].fillna(Military_expenditure_of_GDP)
tabla_indicadores2['General government final consumption expenditure (current US$)'] = tabla_indicadores2['General government final consumption expenditure (current US$)'].fillna(General_government_final_consumption_expenditure_current_US)
tabla_indicadores2['Population growth (annual %)'] = tabla_indicadores2['Population growth (annual %)'].fillna(Population_growth_annual)
tabla_indicadores2['Access to electricity (% of population)'] = tabla_indicadores2['Access to electricity (% of population)'].fillna(Access_to_electricity_of_population)
tabla_indicadores2['Food exports (% of merchandise exports)'] = tabla_indicadores2['Food exports (% of merchandise exports)'].fillna(Food_exports_of_merchandise_exports)
tabla_indicadores2['Food production index (2014-2016 = 100)'] = tabla_indicadores2['Food production index (2014-2016 = 100)'].fillna(Food_production_index_2014_2016_100)
tabla_indicadores2['Households and NPISHs final consumption expenditure (% of GDP)'] = tabla_indicadores2['Households and NPISHs final consumption expenditure (% of GDP)'].fillna(Households_and_NPISHs_final_consumption_expenditure_of_GDP)

```
