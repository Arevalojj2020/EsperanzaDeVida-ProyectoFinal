```python
def getDataFrameDatabank(paises_lista, diccionario_de_datos):
    dataset = np.array([]) 

    for i in diccionario_de_datos.items():
        test = wb.data.fetch(i[0], paises_lista)  # Obtener los datos del indicador económico para los países dados
        time.sleep(1)  # Esperar 1 segundo (pausa)
        lista_de_paises = []  # Lista para almacenar los nombres de los países
        lista_de_años = []  # Lista para almacenar los años
        lista_de_indicadores = []  # Lista para almacenar los datos

        for a in test:
            year = int(a['time'][2::])
            if year >= 1990 and year <= 2021:
                lista_de_paises.append(a['economy'])  # Agregar el nombre del país a la lista
                lista_de_años.append(year)  # Agregar el año a la lista
                lista_de_indicadores.append(a['value'])  # Agregar el valor del indicador económico a la lista

        add = {'title': i[1], 'economy': lista_de_paises, 'time': lista_de_años, 'value': lista_de_indicadores}  # Crear un diccionario con los datos del indicador
        dataset = np.append(dataset, add)  # Agregar el diccionario al dataset

    # Validación de las dimensiones de los datos
    for i in range(0, len(dataset)):
        print('dimention of data list: ' + str(dataset[i]['title']))  # Imprimir el título del indicador económico
        print(len(dataset[i]['value']))  # Imprimir la longitud de la lista de valores del indicador

    for i in range(0, len(dataset) - 1):
        if dataset[i]['time'] == dataset[i + 1]['time']:
            print('is equal times')  # Imprimir mensaje si los tiempos son iguales entre indicadores consecutivos
        if dataset[i]['economy'] == dataset[i + 1]['economy']:
            print('is equal countries')  # Imprimir mensaje si los países son iguales entre indicadores consecutivos

    # Exportar los datos a un DataFrame
    tabla_indicadores = pd.DataFrame()
    tabla_indicadores['country'] = dataset[0]['economy']  # Crear columna 'country' en el DataFrame y asignar los nombres de los países del primer indicador económico
    tabla_indicadores['year'] = dataset[0]['time']  # Crear columna 'year' en el DataFrame y asignar los años del primer indicador económico

    for i in range(0, len(dataset)):
        years = dataset[i]['time']
        values = dataset[i]['value']
        filtered_years = []
        filtered_values = []

        for j in range(len(years)):
            year = int(years[j])
            if year >= 1990 and year <= 2021:
                filtered_years.append(years[j])
                filtered_values.append(values[j])

        tabla_indicadores[dataset[i]['title']] = filtered_values  # Crear columna para cada indicador económico en el DataFrame

    return tabla_indicadores
```
